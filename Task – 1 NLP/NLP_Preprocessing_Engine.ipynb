Q1. Difference between "Love" and "love" in NLP:
In NLP, "Love" and "love" are treated as different words because of case sensitivity. This can increase vocabulary size unnecessarily. So, we usually convert text to lowercase to treat them as the same word.

Q2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset will contain many unnecessary words like "is", "the", "and". This increases noise and can reduce model performance.

Q3. Two cases where removing stopwords is harmful:
1. In sentiment analysis, words like "not" are important. Removing them can change meaning (e.g., "not good" becomes "good").
2. In question answering systems, words like "what", "where" are important and should not be removed.

Q4. Difference between stemming and lemmatization:
Stemming removes suffixes to get root words but may produce incorrect words (e.g., "running" → "run").
Lemmatization converts words into meaningful base forms using vocabulary (e.g., "better" → "good").

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)

    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Remove emojis & special characters
    text = re.sub(r"[^\w\s]", "", text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (<=2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
all_tokens = []

for s in sentences:
    tokens, clean = preprocess_text(s)
    all_tokens.extend(tokens)

    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned:", clean)

    print("Total tokens:", len(tokens))
    print("Unique tokens:", len(set(tokens)))

    if len(tokens) > 0:
        avg_len = sum(len(t) for t in tokens) / len(tokens)
    else:
        avg_len = 0

    print("Avg token length:", round(avg_len, 2))
    print("-" * 50)
```

```python
counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
test_cases = ["", "😍😍😍", "123456"]

for t in test_cases:
    tokens, clean = preprocess_text(t)
    print("Input:", t)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-"*30)
```

Which sentence had most noise?
"Get 100% FREE access now!!!" and "Win $$$ now!!! Limited offer!!!" had most noise due to symbols, numbers, and formatting.

Which sentence retained most meaningful tokens?
"I am not happy with this" retained meaningful tokens because it contains important sentiment words like "not" and "happy".
